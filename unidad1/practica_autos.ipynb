# Realizar un Histograma de precios de los automoviles

```python
n = len(data)
k = int (1+ math.log2(n))

print ("Valor de n=", n, "\nK o numero de intervalos o clases=", k)
plt.hist(data['price'], bins=8)
plt.xlabel('valores')
plt.ylabel('precio de los automoviles')
plt.title('histograma de precios de autos')
```

en el histograma podemos observar como el precio de los autos mas vendidos va desde los 5000 a 10000 dolares

# Realizar un diagrama de caja y bigotes que muestre la arelacion estadistica entre la columna del precio del automovil (price) y tipo de vehiculo (body style)

```python
sns.boxplot(x = "body-style", y = "price", data=data)
plt.title("Relaci√≥n Estadistica del precio del autom√≥vil y el tipo de veh√≠culo")
plt.xlabel("Estilo del Cuerpo del Autom√≥vil")
plt.ylabel("Precio del Autom√≥vil")
```

con este diagrama de caja y bigote podemos ver la relacion que tiene el precio con los tipos de autos mas vndidos podemos ver como los autos mas baratos vendidos son de tipo hatchback, sedan y wagon y como los autos mas caros vendidos son de tipo hardtop y convertible

# Realizar un diagrama de caja y bigotes que muestre la arelacion estadistica entre la columna del precio del automovil (price) y tipo traccion ruedas (drive wheels)

```python
sns.boxplot(x = "drive-wheels", y = "price", data=data)
plt.title("Relaci√≥n Estadistica del precio del autom√≥vil y el tipo de tracci√≥n de ruedas del veh√≠culo.")
plt.xlabel("Tipo de Tracci√≥n")
plt.ylabel("Precio del Autom√≥vil")
```

con este diagrama podemos observar como los autos mayormente tienen una tracion trasera y como sus precios son desde los 13000 dolares hasta los 25000 dolares y lo autos co tracion delantera y 4x4 van desde los 5000 dolares hasta los 15000 dolares


#  Realizar un diagrama de caja y bigotes que muestre la arelacion estadistica entre la columna del precio del automovil (price) y la ubicacion del motor (engine-location)

```python
sns.boxplot(x = "engine-location", y = "price", data=data)
plt.title("Relaci√≥n Estadistica del precio del autom√≥vil y la ubicaci√≥n del motor.")
plt.xlabel("Ubicaci√≥n del motor")
plt.ylabel("Precio del Autom√≥vil")
```

con este diagrama podemos observar como los autos mas caros tienen el motor en la parte trasera mientras que los autos mas baratos cuentan con el motor en la parte de adelante

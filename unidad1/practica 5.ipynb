## ejemplo 3 histogramas de dos grupos
1. en este ejemplo creamos una lista de 100 numeros reales que represetan la produccion diaria de 100 telares(mediads en metros)
2. generamos otra lista con la produccion de 100 telares para una segunda fabrica
3. con estas listas de valores se pretende determinar las medias de tendencia central es decir la medi, la moda y la mediana y la desviacion estandar
4. graficar un histograma para cada fabrica

```python
import random
```

```python
fabrica1=[]
for i in range(100):
  numero_aletorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aletorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
  numero_aletorio=round(random.uniform(50,100),2)
  fabrica2.append(numero_aletorio)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("histograma fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("histograma fabrica 2")
```

```python
media =mean (fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)
print("valores de la fabrica 1")
print("la media es",media)
print("la mediana es",mediana)
print("la moda es",moda)
```

```python
media =mean (fabrica2)
mediana=median(fabrica2)
moda=mode(fabrica2)
print("valores de la fabrica 2")
print("la media es",media)
print("la mediana es",mediana)
print("la moda es",moda)
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2)
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
media1 =mean (fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
plt.hist(fabrica1, bins=10, alpha=0.2,color="b", edgecolor="black")
plt.axvline(media1, color="b", linestyle="dashed", linewidth=2, label="media")
plt.axvline(mediana1, color="r", linestyle="dashed", linewidth=2, label="mediana")
plt.axvline(moda1, color="y", linestyle="dashed", linewidth=2, label="moda")
plt.xlabel=("valor")
plt.ylabel=("frecuencia")
plt.title("producion de la fabrica 1")
plt.legend()
plt.show()
```

```python
media2 =mean (fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
plt.hist(fabrica2, bins=10, alpha=0.2,color="b", edgecolor="black")
plt.axvline(media2, color="b", linestyle="dashed", linewidth=2, label="media")
plt.axvline(mediana2, color="r", linestyle="dashed", linewidth=2, label="mediana")
plt.axvline(moda2, color="y", linestyle="dashed", linewidth=2, label="moda")
plt.xlabel=("valor")
plt.ylabel=("frecuencia")
plt.title("producion de la fabrica 2")
plt.legend()
plt.show()
```

#DISPERCION DE LOS DATOS


```python
DESV1=stdev(fabrica1)
print("la desviacion estandar de la fabrica 1 es de \n", DESV1)
```

```python
DESV2=stdev(fabrica2)
print("la desviacion estandar de la fabrica 2 es de \n", DESV2)
```

```python
media1 =mean (fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
plt.hist(fabrica1, bins=10, alpha=0.2,color="b", edgecolor="black")
plt.axvline(media1, color="b", linestyle="dashed", linewidth=2, label=f"media={media1:.2f}")
plt.axvline(mediana1, color="r", linestyle="dashed", linewidth=2, label=f"mediana={mediana1:.2f}")
plt.axvline(moda1, color="y", linestyle="dashed", linewidth=2, label=f"moda={moda1:.2f}")

plt.axvline(media1 - DESV1, color="black", linestyle="dashed", linewidth=2, label=f"desviacion estandar={DESV1:.2}")
plt.axvline(media1+DESV1, color="black", linestyle="dashed", linewidth=2)
plt.xlabel=("valor")
plt.ylabel=("frecuencia")
plt.title("producion de la fabrica 1")
plt.legend()
plt.show()
```

```python
media2 =mean (fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
plt.hist(fabrica2, bins=10, alpha=0.2,color="b", edgecolor="black")
plt.axvline(media2, color="b", linestyle="dashed", linewidth=2, label=f"media={media2:.2f}")
plt.axvline(mediana2, color="r", linestyle="dashed", linewidth=2, label=f"mediana={mediana2:.2f}")
plt.axvline(moda2, color="y", linestyle="dashed", linewidth=2, label=f"moda={moda2:.2f}")

plt.axvline(media2 - DESV2, color="black", linestyle="dashed", linewidth=2, label=f"desviacion estandar={DESV2:.2}")
plt.axvline(media2+DESV2, color="black", linestyle="dashed", linewidth=2)
plt.xlabel=("valor")
plt.ylabel=("frecuencia")
plt.title("producion de la fabrica 2")
plt.legend()
plt.show()
```

#tablas de frecuencia


```python
edades=[]
for i in range(100):
  numero_al=random.randint(18,70)
  edades.append(numero_al)
print(edades)
```

```python
import pandas as pd
```

```python
df_edades = pd.DataFrame(edades, columns=["edad"])
df_edades
```

```python
##calcular la tabla de distribucion de frecunecias

tablafre = df_edades["edad"].value_counts().reset_index()
tablafre.columns = ["edad", "frecuencia"]
tablafre = tablafre.sort_values(by="edad")
tablafre
```

```python
tablafre["frecuencia relativa"]=tablafre["frecuencia"]/100
```

```python
tablafre
```

```python
tablafre["frecuencia porcentual"] = (tablafre["frecuencia"] / tablafre["frecuencia"].sum()) * 100

tablafre
```

```python
#frecuencia acumulada
tablafre["frecuencia acumulada"] = tablafre["frecuencia"].cumsum()
tablafre
```

#regla de sturges


```python
import math
```

```python
df_edades_s=pd.DataFrame({"edades":edades})
df_edades_s
```

```python
n=len(edades)
k=int(1 + math.log2(n))
print("valor de b=",n,"k o numero de intervalos",k)
```

```python
plt.hist(edades, alpha=0.5, rwidth=0.95)
plt.title("histograma de edades basado en valores de matplotlib")
```

```python
tablades = pd.cut(df_edades_s["edades"], bins = k).value_counts().reset_index()
tablades
```

```python
tablades.columns=["clase", "frecuencias"]
tablades
```

```python
tablades=tablades.sort_values(by="clase")
tablades
```

```python
tablades["frec relativa"]= tablades["frecuencias"]/n
tablades["frec porcentual"]= tablades["frecuencias"]/n*100
tablades["frec acunulada"]= tablades["frecuencias"].cumsum()
```
